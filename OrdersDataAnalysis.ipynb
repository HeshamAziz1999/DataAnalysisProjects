```python
import pandas as pd
```

```python
df = pd.read_csv("Orders.csv",skiprows=4)
df
```

```python
df.dropna(how = 'all', inplace= True)
df
```

```python
df["City"]= df["City"].str.title()
df
```

```python
df['Branch']= df['Branch'].str.upper()
df['Email']= df['Email'].str.lower()
df
```

```python
df['Phone Number']= df['Phone Number'].str.replace('Tel: ', '')
df
```

```python
df['Discount'] = df['Discount'].round(0)
df
```

```python
df['Total_Price'] = df['Quantity'] * df['Unit Price']
df['Cost'] = df['Total_Price'] - df['Discount'].fillna(0)    
df
```

```python
df.insert(9, 'Customer Full Name',df['Customer First Name'].str.title() + " "+ df['Customer Last Name'].str.title())
```

```python
df
```

```python
newCoulms = df['Branch'].str[:2]
df.insert(5 ,'BrCode', newCoulms)
```

```python
df
```

```python
df.loc[df['Status']==True, 'Order Status'] = 'Done'
df.loc[df['Status']==False, 'Order Status'] = 'Pending'
df.loc[(df['Status']==False) & (df['Order Date']<='2023-01-01'), 'Order Status'] = 'Canceled'
df
```

```python
def level(row):
    if row > 1000:
        return 'High'
    else:
        return 'Low'
df['Cost Level'] = df['Cost'].apply(level)
df
```

```python
import numpy as np
```

```python
df['Discount Description'] = np.where(df['Discount'].isnull(), 'No Discount', (df['Discount']/df['Cost']*100).round(2))
df
```
