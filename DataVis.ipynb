## This is PLot to show Which country Have Most Orders

```python
OrdersByCountry= df['Country'].value_counts()
```

```python
OrdersByCountry.plot(kind='bar', title = 'OrdersByCountry')
```

## Saudi Arabia Have most Orders 
### Through this chart, the company can determine which countries it wants to increase sales in.


```python
Cat_Cost_Total = df.groupby('Category')['Total Cost'].sum()
Cat_Cost_Total
```

### this Chart show which Categroy Have Biggest Cost

```python
Cat_Cost_Total.plot(kind='barh', color='green', xlabel='Total Cost(Millions)', title= 'Categroy By Total Cost')
```

## Laptop & Digital Camera Have biggest Total Cost 
#### We need more Insight to take decision 

```python
Cat_Quan_TotalCost = df.groupby('Category')[['Quantity', 'Total Cost']].sum()

Cat_Quan_TotalCost
```

# This Chart Show Total Cost & Quantity By Categroy
##### Through this statement we can make decisions such as: The categories that are more expensive for the company and their quantity is small, the company can reconsider the pricing role

```python
Sub_x=Cat_Quan_TotalCost.plot(kind='bar',y='Total Cost',ylabel= 'Total Cost (Millions)', color= 'g')
Cat_Quan_TotalCost.plot(kind='line',y='Quantity',ylabel= 'Quantity', color= 'r',secondary_y=True, ax=Sub_x , rot=45)
```
